```python
import yaml
import csv

with open('source/schema.yml', 'r', encoding='utf-8') as f:
    example = yaml.safe_load(f)

example
```

```python
with open('all_tables.yaml', 'r') as f:
    tables = yaml.safe_load(f)



# tables_dict = {'description': 'INCPT table', 'name': 'INCPT', 'columns':[{'description': '', 'name': 'DSCAMT'}]}

list_table_dict = []
len(tables)
```

```python
for i in tables:   
    with open(f'source_tables_csv/{i}.csv', 'r', encoding='utf-8') as f:
        table= csv.reader(f.readlines())
        columns = [col for col in table]
        tables_desc = columns[:columns.index([])]
        columns = columns[columns.index([])+1:]
        while 1:
            try:
                columns.remove([])
            except:
                break
        if not all((len(item) == len(columns[0])) for item in columns):
            raise ValueError
        list_column_dict = [{'name':column[1], 'description': column[3]} for column in columns[1:]]
        if len(tables_desc) == 1:
            tables_dict = {'description': None, 'name': i, 'columns':list_column_dict}
        else:
            tables_dict = {'description': tables_desc[1][2], 'name': i, 'columns': list_column_dict}
            
    list_table_dict.append(tables_dict)

len(list_table_dict)
```

```python
example['sources'][0]['tables'] = list_table_dict
```

```python
with open('source/schema.yml', 'w', encoding='utf-8') as f:
    yaml.safe_dump(example, f, default_flow_style=False, allow_unicode=True)
```
